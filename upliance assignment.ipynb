```python
# Example: Aggregating orders and revenue for each dish
popular_dishes = data.groupby('Dish Name').agg(
    Total_Orders=('Order ID', 'count'),  # Count of orders
    Total_Revenue=('Amount (USD)', 'sum')      # Sum of revenue
).reset_index()
```

```python
# # Sort dishes by popularity (orders) or revenue
popular_dishes = popular_dishes.sort_values(by='Total_Orders', ascending=False)
```

```python
# # Display the most popular dishes
print(popular_dishes.head())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Take the top 5 most popular dishes (by orders)
top_dishes = popular_dishes.head(5)

# Plot a bar chart for total orders
plt.figure(figsize=(10, 6))
sns.barplot(data=top_dishes, x='Dish Name', y='Total_Orders', palette='viridis')
plt.title('Top 5 Most Popular Dishes', fontsize=16)
plt.xlabel('Dish Name', fontsize=12)
plt.ylabel('Total Orders', fontsize=12)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
sns.barplot(data=data, x='Location', y='Total Orders', palette='viridis')
plt.title('Total Orders by Location', fontsize=16)
plt.xlabel('Location', fontsize=12)
plt.ylabel('Total Orders', fontsize=12)
plt.xticks(rotation=45, fontsize=10)
plt.tight_layout()
plt.show()
```

```python
# Group data by location
orders_by_location = data.groupby('Location').agg(
    Total_Orders=('Order ID', 'count'),  # Total number of orders
    Avg_Orders_Per_User=('Order ID', 'mean')  # Average orders per user
).reset_index()

# Sort by total orders
orders_by_location = orders_by_location.sort_values(by='Total_Orders', ascending=False)

# Display results
print(orders_by_location.head())
```
